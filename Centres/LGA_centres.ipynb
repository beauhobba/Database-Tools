```python

f = open("./LGA_centres.txt", "w", encoding="utf-8")

f.write('LATITUDE\tLONGITUDE\tLGA_CODE22\tLGA_NAME22\tSTE_CODE21\tSTE_NAME21\tAREASQKM\tSHAPE_Area\tSHAPE_Leng\n')

geos = shape['geometry']
LGA_CODE20 = shape['LGA_CODE22']
LGA_NAME20 = shape['LGA_NAME22']
STE_CODE16 = shape['STE_CODE21']
STE_NAME16 = shape['STE_NAME21']
AREASQKM20 = shape['AREASQKM']
SHAPE_Area = shape['SHAPE_Area']
SHAPE_Leng = shape['SHAPE_Leng']

# input(len(LGA_CODE20))

no_centroids = np.array([])
for geo, lga_code, lga_name, ste_code, ste_name, area, sh_area, sh_len in zip(geos, LGA_CODE20, LGA_NAME20, STE_CODE16, STE_NAME16, AREASQKM20, SHAPE_Area, SHAPE_Leng):
    try:
        centroid = geo.centroid 
        lat = centroid.y
        lon = centroid.x
        # input(f"{lat}\t{lon}\t{lga_code}\t{ste_code}\t{ste_name}\t{area}\t{sh_area}\t{sh_len}\n")
        f.write(f"{lat}\t{lon}\t{lga_code}\t{lga_name}\t{ste_code}\t{ste_name}\t{area}\t{sh_area}\t{sh_len}\n")
    except Exception as ex:
        print(f"{lga_code} {ex}")
        no_centroids = np.append(no_centroids, lga_code)
f.close()

# Now write in the values with no centroids
f = open("./no_LGA_centres.txt", "w", encoding="utf-8")
f.write('LGA_CODE20\n')
for code in no_centroids:
    f.write(f"{code}\n")
f.close()

        
    
```
