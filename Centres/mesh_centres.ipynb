```python

f = open("./mesh_centres.txt", "w", encoding="utf-8")
f.write('MB_CODE21\tLATITUDE\tLONGITUDE\tMB_CAT21\tLOCI_URI21\tAUS_CODE21\tGCC_CODE21\tGCC_NAME21\tAUS_NAME21\tCHG_LBL21\tCHG_FLAG21\tSTE_NAME21\tSTE_CODE21\tAREASQKM21\tSA1_CODE21\tSA2_CODE21\tSA2_NAME21\tSA3_CODE21\tSA3_NAME21\tSA4_CODE21\tSA4_NAME21\n')

geos = shape['geometry']
MB_CODE = shape['MB_CODE21']
MB_CAT = shape['MB_CAT21']
LOCI = shape['LOCI_URI21']
AUS_CODE21 = shape['AUS_CODE21']
GCC_CODE = shape['GCC_CODE21']
GCC_NAME = shape['GCC_NAME21']
AUS_NAME = shape['AUS_NAME21']
CHG_LBL = shape['CHG_LBL21']
CHG_FLAG = shape['CHG_FLAG21']
STE_NAME = shape['STE_NAME21']
STE_CODE = shape['STE_CODE21']
AREASQK = shape['AREASQKM21']
SA1 = shape['SA1_CODE21']
SA2 = shape['SA2_CODE21']
SA2_NAME = shape['SA2_NAME21']
SA3_CODE = shape['SA3_CODE21']
SA3_NAME = shape['SA3_NAME21']
SA4_CODE = shape['SA4_CODE21']
SA4_NAME = shape['SA4_NAME21']

no_centroids = np.array([])
for geo, name, cat, loci, aus_code, gcc_code, gcc_name, aus_name, chg_lbl, chg_flag, ste_name, ste_code, areasqk, sa1, sa2, sa2_name, sa3_code, sa3_name, sa4_code, sa4_name in zip(geos, MB_CODE, MB_CAT, LOCI,AUS_CODE21, GCC_CODE, GCC_NAME, AUS_NAME, CHG_LBL, CHG_FLAG, STE_NAME, STE_CODE, AREASQK, SA1, SA2, SA2_NAME, SA3_CODE, SA3_NAME, SA4_CODE, SA4_NAME):
    try:
        centroid = geo.centroid 
        lat = centroid.y
        lon = centroid.x
        f.write(f"{name}\t{lat}\t{lon}\t{cat}\t{loci}\t{aus_code}\t{gcc_code}\t{gcc_name}\t{aus_name}\t{chg_lbl}\t{chg_flag}\t{ste_name}\t{ste_code}\t{areasqk}\t{sa1}\t{sa2}\t{sa2_name}\t{sa3_code}\t{sa3_name}\t{sa4_code}\t{sa4_name}\n")
    except Exception as ex:
        print(f"{name} {ex}")
        no_centroids = np.append(no_centroids, name)
f.close()

# Now write in the values with no centroids
f = open("./no_mesh_centres.txt", "w", encoding="utf-8")
f.write('MB_CODE21\n')
for code in no_centroids:
    f.write(f"{code}\n")
f.close()

        
    
```
